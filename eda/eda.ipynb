## Anomaly Detection Using Visualisation

```python
# Univariate Analysis 

for i in range(len(df.columns)-1):
    sns.countplot(data=df, x= df.columns[i])
    plt.show()
    
```

```python
# Bivariate Analysis 
for i in range(len(df.columns)- 1):
    sns.lineplot(data=df, x=df.columns[i], y=df.columns[i+1])
    plt.show()
    
```
