# Boiler System Dataset: Column Descriptions

- **Datetime**: The timestamp.
- **BOI_FLOW_1**: Water flow rate through Boiler 1 (GPM).
- **BOI_FLOW_2**: Water flow rate through Boiler 2 (GPM).
- **BOI_GAS_CSUM_1**: Cumulative gas consumption of Boiler 1 (kW).
- **BOI_GAS_CSUM_2**: Cumulative gas consumption of Boiler 2 (kW).
- **BOI_RW_TEMP_1**: Return water temperature to Boiler 1 (°F).
- **BOI_RW_TEMP_2**: Return water temperature to Boiler 2 (°F).
- **BOI_STA_1**: Operational status of Boiler 1 (1 = ON, 0 = OFF).
- **BOI_STA_2**: Operational status of Boiler 2 (1 = ON, 0 = OFF).
- **BOI_SW_TEMP_1**: Supply water temperature from Boiler 1 (°F).
- **BOI_SW_TEMP_2**: Supply water temperature from Boiler 2 (°F).
- **HWL_DP**: Differential pressure in the hot water loop (in H₂O).
- **HWL_DPSPT**: Setpoint for differential pressure in the hot water loop (in H₂O).
- **HWL_RW_TEMP**: Return water temperature in the hot water loop (°F).
- **HWL_SW_TEMP**: Supply water temperature in the hot water loop (°F).
- **OA_TEMP**: Outdoor air temperature (°F).
- **PM_POW_1**: Power consumption of Pump 1 (kW).
- **PM_POW_2**: Power consumption of Pump 2 (kW).
- **PM_SPD_1**: Speed ratio of Pump 1 (0 to 1).
- **PM_SPD_2**: Speed ratio of Pump 2 (0 to 1).
- **PM_STA_1**: Operational status of Pump 1 (1 = ON, 0 = OFF).
- **PM_STA_2**: Operational status of Pump 2 (1 = ON, 0 = OFF).
- **SEC_POW**: Secondary loop power consumption (kW).
